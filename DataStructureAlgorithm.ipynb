# Sorting

```python
List = [4, 5, 2, 10, 23, 100]
```

```python
# Bubble Sort
# essense: every time compare the adjacent two elements and sort them consecutively 
# --> each occurence will find the largest value and place it at right most
def bubble_sort(List):
    for occ in range(len(List)):
        print(occ)
    # repeat the occurence     
        for i in range(len(List) - 1 - occ):
            if List[i] > List[i+1]:
                List[i], List[i+1] = List[i+1], List[i]
    return List
```

```python
bubble_sort(List)
```

```python
%%timeit
bubble_sort(List)
```

time complexity:
- 1st occ --> n-1
- 2nd occ --> n-2
...
- last occ --> 1

(O) = n-1 + n-2 + n-3 + ... + 1 = {[(n-1) + 1] * (n-1)}/2 = (n^2-n)/2 = n^2

```python
# insertion sort
# essence: assume the list at front is sorted
def insertion_sort(List):
    for i in range(1, len(List)):
        key = List[i]
        print(f"when key = {key}")
        j = i-1
        print(f"j={j}")
        while (key < List[j]) and j >= 0:
            List[j+1] = List[j]
            j-=1
        List[j+1] = key
    return List
```

```python
insertion_sort(List)
```

```python
%%timeit
insertion_sort(List)
```

```python
# selection sort
# essense: search the smallest element and swap with 1st element
def selection_sort(List):
    for i in range(len(List) - 1):
        print(f"when i = {i}")
        index = i
        for j in range(i+1, len(List)):
            if List[index] > List[j]:
                index = j
        List[i], List[index] = List[index], List[i]
    return List            
```

```python
selection_sort(List)
```

- (n-1)th --> 1

```python
# Merge sort 
# essense: devide the list into single elements, then merge them back as two sorted lists
def merge_sort(List):
    # First to define the stopping criteria for recursive, if no element, then do nothing
    if len(List) <= 1:
        return List
    # if > 1 element, then recursive
    mid = len(List)//2
    left = List[:mid]
    right = List[mid:]
    merge_sort(left)
    merge_sort(right)
    # now left and right are two sorted lists, so we need to merge them
    print(left, right)
    i, j, k = 0, 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            List[k] = left[i]
            k += 1
            i += 1
        else:
            List[k] = right[j]
            k += 1
            j += 1
    # once while loop is stopped, one of the i and j must be outranged, what we need now is to continue placing for the rest of elements
    while i < len(left):
        List[k] = left[i]
        k += 1
        i += 1
    while j < len(right):
        List[k] = right[j]
        k += 1
        j += 1
    return List
```

```python
# partition return the index of the pivot, and perform the "sort" for leftside and rightside of the List
def partition(List, start, end):
    # this defines if the pivot is from start or end
    pivot_index = start
    pivot = List[pivot_index]
    print(f"pivot = {pivot}")
    
    while start < end:
        # find the value > pivot on left
        while start < len(List) and List[start] <= pivot:
            start += 1
        # find the value <= pivot on the right
        while List[end] > pivot:
            end -= 1

        if start < end:
            List[start], List[end] = List[end], List[start]
    
    List[end], List[pivot_index] = List[pivot_index], List[end]
    print(f"end = {end}")
    return end



# quick sort
# essense: find a pivot element(can be first, last or any other element)
# apply partition function to sort the List, and then return the actual index of the pivot
# recursively quick sort both side
def quick_sort(List, start, end):
    # define stopping criteris
    if start >= end:
        return List
    # let say 2 elements in the list
    p = partition(List, start, end)
    quick_sort(List, start, p-1)
    quick_sort(List, p+1, end)
    return List
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        # next is also a node type
        self.next = None

    def __repr__(self):
        return self.data

class LinkedList:
    # LinkedList just track the head node
    def __init__(self):
        # head is a node type
        self.head = None
        
    # traverse an iterator
    def __iter__(self):
        print("iter")
        node = self.head
        while node is not None:
            yield node
            node = node.next
    
    def add_first(self, node):
        node.next = self.head
        self.head = node
    
    def add_last(self, node):
        if self.head == None:
            self.head = node
            
        for current_node in self:
            pass
        current_node.next = node
        
    
    def __repr__(self):
        node = self.head
        nodes = []
        while node is not None:
            nodes.append(node.data)
            node = node.next
        nodes.append("None")
        return " -> ".join(nodes)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        # next is also a node type
        self.next = None

    def __repr__(self):
        return self.data

class LinkedList:
    # LinkedList just track the head node
    def __init__(self):
        # head is a node type
        self.head = None
        self.top = None
    
    def add_first(self, node):
        node.next = self.head
        self.head = node
    
    def add_last(self, node):
        self.top.next = node
        self.top = node
        
    def __repr__(self):
        node = self.head
        nodes = []
        while node is not None:
            nodes.append(node.data)
            node = node.next
        nodes.append("None")
        return " -> ".join(nodes)
```

```python
llist = LinkedList()

first_node = Node("a")
second_node = Node("b")
third_node = Node("c")
first_node.next = second_node
second_node.next = third_node

llist.head = first_node
llist.top = third_node
```

```python
class PowTwo:
    """Class to implement an iterator
    of powers of two"""

    def __init__(self, max=0):
        self.max = max

    def __iter__(self):
        # return an iterator object, which is itself
        self.n = 0
        print("iter")
        return self
    
    def __next__(self):
        # iterator object must have __next__ method
        if self.n <= self.max:
            print("__next__.n:", self.n)
            result = 2 ** self.n
            self.n += 1
            return result
        else:
            raise StopIteration
```

```python
# create an object
numbers = PowTwo(3)
```

```python
# create an iterator
i = iter(numbers)
```

```python
isinstance(numbers, collections.Iterable)
```
